```python
df.head()
```

```python
print(df[~df['Release_Date'].str.match(r'\d{1,2}-\d{1,2}-\d{4}', na=False)])
```

```python
# removing non date values Rows
df = df[df['Release_Date'].str.match(r'\d{1,2}-\d{1,2}-\d{4}', na=False)]
```

```python
# converting the data values to specific format
df['Release_Date'] = pd.to_datetime(df['Release_Date'], format='%d-%m-%Y', dayfirst=True)
```

```python
df.info()
```

```python
# converting vote count to Integer and vote average to float
df['Vote_Count'] = df['Vote_Count'].astype(float)
df['Vote_Average'] = df['Vote_Average'].astype(float)
```

```python
df.info()
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
```

```python
df.head()
```

```python
# Droping the columns
cols =['Overview','Original_Language','Poster_Url']
df.drop(columns=cols,inplace=True)
```

```python
df.head()
```

```python
def categorize_column(df,col,labels):
  edges = [df[col].describe()['min'],
           df[col].describe()['25%'],
           df[col].describe()['50%'],
           df[col].describe()['75%'],
           df[col].describe()['max']]
  df[col] = pd.cut(df[col],edges,labels=labels,duplicates='drop')
  return df

```

```python
labels =['not popular','Below_average','average','popular']
df = categorize_column(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# splitting the genre column and exploding it for only one genre for each row.
df['Genre'] = df['Genre'].str.split(", ")
df = df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
# casting column into category
df['Genre'] = df['Genre'].astype('category')
df['Genre'].dtypes
```
